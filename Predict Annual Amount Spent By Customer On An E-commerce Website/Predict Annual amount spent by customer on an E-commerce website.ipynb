```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### To read data from csv file

```python
dataset = pd.read_csv("Ecommerce Customers.csv")
```

```python
dataset.head()
```

### Assume, x is independent variable and y is dependent variable

```python
x = dataset.iloc[:,3:7].values #columns 3 to 6 are selected
y = dataset.iloc[:,7:].values #last col i.e.7th is selected
```

```python
x
```

```python
y
```

```python
# X = dataset.iloc[:,3:4].values  # we got scattered data in 3rd col
# X = dataset.iloc[:,4:5].values  # we got scattered data in 4th col
# X = dataset.iloc[:,5:6].values  # we got scattered data in 5th col
X = dataset.iloc[:,6:7].values  # we got linear data in 6th col

#visualize
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.scatter(X,y,color="blue")
```

### Split data into train and test data

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
#train_test_split (independent, dependent, 20% split i.e.20% of test data, random  arrengement of selection)
```

```python
x_test
```

```python
y_test
```

### Use linear regression model

##### we've multiple independent var and 1 dependent var, Hence, Multiple Linear Regression Model

```python
from sklearn.linear_model import LinearRegression
```

```python
regression = LinearRegression()
```

#### Apply model to trained dataset

```python
regression.fit(x_train, y_train)
```

```python
y_predict = regression.predict(x_test)
```

```python
y_predict
```
